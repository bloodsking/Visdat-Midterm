```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('spotify_songs.csv')
df.head()
```

```python
df.info()
```

```python
df['playlist_subgenre'].unique()
```

```python
df.corr()
```

```python
df.drop(['track_id','track_album_id','key','playlist_id', 'playlist_name', 'speechiness', 'mode', 'playlist_subgenre'], axis=1, inplace=True)
df
```

```python
df = df.drop_duplicates(subset=None, keep='first', inplace=False)
```

```python
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'])
```

```python
def milidetik_ke_detik(milidetik):
    return milidetik / 1000
```

```python
df['duration_sec'] = df['duration_ms'].apply(milidetik_ke_detik)
```

```python
df['song_category'] = pd.cut(df['duration_sec'], bins=[0, 120, 220, float('inf')], labels=['Short', 'Average', 'Long'])
```

```python
df['track_artist'] = df['track_artist'].replace('Joey Bada$$', 'Joey Badass')
```

```python
df
```

```python
df = df.drop('duration_ms', axis=1)
df
```

## Visualisasi 1

Visualitation Total Released Song from 2014 to 2019

```python
df1 = df[(df['track_album_release_date'].dt.year >= 2010) & (df['track_album_release_date'].dt.year <= 2019)]
```

```python
totalLagu = df1['track_album_release_date'].dt.year.value_counts().sort_index()
```

```python
totalLagu
```

```python
colors = ['royalblue', 'orangered', 'forestgreen', 'gold', 'mediumvioletred', 'darkorange', 'lightseagreen']
```

```python
plt.figure(figsize=(10, 6))
totalLagu.plot(kind='bar', color=colors)
plt.title('Jumlah Lagu yang Dirilis dari Tahun 2014 hingga 2019')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Lagu')

plt.annotate('Ramainya TikTok', xy=(9, totalLagu[2019]), xytext=(6, totalLagu[2019] + 3),
             arrowprops=dict(arrowstyle='->'), fontsize=12, color='red')
plt.grid()
```

# Visualisasi 2

Visualitation Track Popularity Dengan Album Release Terhadap Genre

```python
df2 = df[(df['track_album_release_date'].dt.year >= 1950) & (df['track_album_release_date'].dt.year <= 2019)]
```

```python
df2 = df2[df2['track_popularity'] > 75]
```

```python
plt.figure(figsize=(15, 8))
scatter = plt.scatter(
    df2['track_album_release_date'],
    df2['track_popularity'],
    s=df2['energy']*1000, 
    c=df2['energy'],  
    cmap='viridis',
    alpha=0.7
)

plt.title('Transisi Song Energy dengan Popularity (1950-2020)')
plt.xlabel('Track Album Release Date')
plt.ylabel('Track Popularity')
plt.grid(True)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator(base=5))

cb = plt.colorbar(scatter, label='Energy')
plt.show()
```

# Visualisasi 3

Visualitation Song Length Category dari 1950 sampai 2019

```python
df3 = df[(df['track_album_release_date'].dt.year >= 1950) & (df['track_album_release_date'].dt.year <= 2019)]
```

```python
df3 = df3[df3['track_popularity'] > 75]
```

```python
songKategori = df3['song_category'].value_counts()
```

```python
colors_list = ['gold', 'lightskyblue', 'lightcoral']
explode_list = [0, 0, 0.1] # ratio for each continent with which to offset each wedge.

plt.figure(figsize=(6, 6))
plt.pie(songKategori, autopct='%1.1f%%', startangle=180,shadow=True,          
                            pctdistance=1.12, 
                            colors=colors_list, 
                            explode=explode_list)
plt.title('Song Categories From (1990-2019)')
plt.legend(songKategori.index, title='Categories', bbox_to_anchor=(1.05, 1), loc='upper right')
```

# Visualisasi 4

Visualitation Artis dengan Danceability Tinggi 

```python
df4 = df[df['danceability'] > 0.8]
```

```python
df4 = df4[(df4['track_album_release_date'].dt.year >= 1980) & (df4['track_album_release_date'].dt.year <= 2019)]
```

```python
highSong = df4['track_artist'].value_counts().head(8)
highSong
```

```python
plt.figure(figsize=(15, 8))
highSong.sort_values().plot(kind='barh', color=plt.cm.Paired(range(len(highSong))))
plt.title('Jumlah Lagu dengan Danceability Tinggi')
plt.xlabel('Jumlah Lagu')
plt.ylabel('Artis')
plt.grid()
```
